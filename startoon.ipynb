```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the text files
with open('/content/Data_1.txt', 'r') as f:
    data_1 = np.array([float(line.strip()) for line in f])

with open('/content/Data_2.txt', 'r') as f:
    data_2 = np.array([float(line.strip()) for line in f])

# Function to find maxima and minima
def find_peaks(data):
    maxima = []
    minima = []
    for i in range(1, len(data) - 1):
        if data[i] > data[i-1] and data[i] > data[i+1]:
            maxima.append((i, data[i]))
        elif data[i] < data[i-1] and data[i] < data[i+1]:
            minima.append((i, data[i]))
    return maxima, minima

# Find peaks for both datasets
maxima_1, minima_1 = find_peaks(data_1)
maxima_2, minima_2 = find_peaks(data_2)

# Plot the results
def plot_peaks(data, maxima, minima, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data, label='Signal')
    plt.scatter(*zip(*maxima), color='red', label='Maxima')
    plt.scatter(*zip(*minima), color='blue', label='Minima')
    plt.title(title)
    plt.xlabel('Index')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot for Data_1
plot_peaks(data_1, maxima_1, minima_1, 'Peaks in Data 1')

# Plot for Data_2
plot_peaks(data_2, maxima_2, minima_2, 'Peaks in Data 2')
```
